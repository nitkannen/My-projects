```python
df_patients = pd.read_csv('PatientInfo.csv')
```

```python
df_patients.head()
```

```python
df_patients.isnull().sum()
```

```python
df_patients.drop(['disease','infection_order','infected_by','contact_number','symptom_onset_date','released_date','deceased_date','global_num'],axis = 1,inplace = True)
```

```python
df_patients.head()
```

```python
len(df_patients)
```

```python
df_patients.isnull().sum()
```

```python
df_patients.dropna(inplace = True)
```

```python
plt.figure(figsize=(10,6))
plt.title('State of patients gender-wise',fontsize = 20)
sns.set_context(context = 'notebook',font_scale = 1.2)
sns.countplot(x = 'state',data = df_patients,hue = 'sex',palette = 'magma')
```

```python
plt.figure(figsize = (30,15))
plt.title('Province-wise patient count',fontsize = 40)
sns.set_context(context = 'notebook',font_scale = 2)
sns.set_style(style = 'whitegrid')
ax= sns.countplot(x = 'province',data = df_patients,)
ax.set_xticklabels(ax.get_xticklabels(),rotation = 45,ha = 'right',fontsize = 25)
```

```python
#This is done to remove the unnecessary s present in age column 
df_patients['age']= (df_patients['age'].apply(lambda x:float(str(x)[:1])*10))
```

```python
plt.figure(figsize = (10,6))
plt.title("Age Distribution of admitted patients",fontsize = 30)
sns.set_style(style = 'darkgrid')
sns.distplot(df_patients['age'],color = 'green')
```

```python
df_patients.head()
```

```python
dead = df_patients[df_patients['state']=='deceased']
rec = df_patients[df_patients['state']=='released']
iso =df_patients[df_patients['state']=='isolated']
```

```python
ax1= plt.figure(figsize = (10,6))
plt.title("Covid consequences for different ages",fontsize = 30)
sns.set_style(style = 'whitegrid')
sns.set_color_codes(palette = 'deep')
sns.kdeplot(dead['age'],shade =True,color = 'r',label = 'Deceased',legend = False)
ax= sns.kdeplot(rec['age'],shade =True,legend = False,label = 'Recovered')
sns.kdeplot(iso['age'],shade = True,legend = False,color = 'y',label = 'isolated')
ax1.legend(loc = 7)
ax.set_xlabel("Age distribution",fontsize = 18)
```

```python
ax = plt.figure(figsize = (10,6))
sns.set_style('whitegrid')
sns.kdeplot(dead[dead['sex']=='female']['age'],label = 'Female',shade = True,legend =False,color = 'b')
ax1= sns.kdeplot(dead[dead['sex']=='male']['age'],label = 'Male',shade = True,legend = False,color = 'r')
ax.legend(loc = 7)
plt.title("Sex-wise deceased",fontsize = 30)
ax1.set_xlabel("Age Distribution")
```

```python
ax = plt.figure(figsize = (10,6))
sns.set_style('whitegrid')
sns.kdeplot(rec[rec['sex']=='female']['age'],label = 'Female',shade = True,legend =False,color = 'b')
ax1= sns.kdeplot(rec[rec['sex']=='male']['age'],label = 'Male',shade = True,legend = False,color = 'r')
ax.legend(loc = 7)
plt.title("Sex-wise recovered",fontsize = 30)
ax1.set_xlabel("Age Distribution")
```
